```python
name = "Park"
age = 30

# f-string 기본 사용
print(f"이름: {name}, 나이: {age}")  # 변수 값을 문자열 안에 직접 삽입

# 소수점 자리수 지정
pi = 3.141592
print(f"{pi:.2f}")  # 소수점 둘째 자리까지 출력 → 3.14

# 자리수(폭) 지정
num = 7
print(f"{num:5}")   # 전체 5칸 확보, 오른쪽 정렬 → '    7'
print(f"{num:05}")  # 빈칸을 0으로 채움 → 00007
```

```python
# 왼쪽 정렬 (<), 오른쪽 정렬 (>), 가운데 정렬 (^)
print(f"{'Python':<10}")  # 왼쪽 정렬
print(f"{'Python':>30}")  # 오른쪽 정렬
print(f"{'Python':0^10}")  # 가운데 정렬
```

```python
# 기본 사용
print("이름: {}, 나이: {}".format("Park", 30))  # {} 순서대로 값 들어감

# 인덱스 사용
print("이름: {0}, 나이: {1}".format("Park", 30))  # 번호로 위치 지정

# 이름 지정
print("이름: {name}, 나이: {age}".format(name="Park", age=30))  
# 변수 이름으로 직접 지정 가능
```

```python
name = "Park"
age = 30

print("이름: %s, 나이: %d" % (name, age))  
# %s = 문자열, %d = 정수

pi = 3.141592
print("%.2f" % pi)  # 소수점 둘째 자리까지 출력 → 3.14
```

```python
name = "python"

print(f"{name:*<15}") # 왼쪽정렬, 빈칸은 *로 채움
print(f"{name:*^15}") # 중앙정렬, 빈칸은 *로 채움
print(f"{name:*>15}") # 오른쪽정렬, 빈칸은 *로 채움
# 변수 사용 가능
w = 15
print(f"{name:*>{w}}") # 오른쪽정렬, 빈칸은 *로 채움
```
