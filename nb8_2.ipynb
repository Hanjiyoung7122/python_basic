```python
try:
    num = int(input("숫자 입력 : ")) ** 2
    print(num)
except:
    print("숫자가 아닙니다.")

print("다음 실행문...")
```

```python
try:
    num = int(input("숫자 입력 : "))
    result = 10 / num
    print(result)

except ValueError:
    print("숫자를 입력하세요.")

except ZeroDivisionError:
    print("0으로 나눌 수 없습니다.")

print("종료")
```

```python
# finally -> 예외 발생 여부와 관계없이 항상 실행된다.
try:
    print(10 / 2)
except:
    print("오류 발생")
finally:
    print("프로그램 종료")
```

```python
try:
    num = int(input("숫자 입력: "))
    result = 100 / num

except ValueError:
    print("숫자를 입력해야 합니다.")

except ZeroDivisionError:
    print("0으로 나눌 수 없습니다.")

else:
    print("계산 결과:", result)

finally:
    print("프로그램 종료")
```
