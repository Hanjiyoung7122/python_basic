```python
def outer():
    def inner():  #함수안에도 함수를 중첩 할 수 있다 (데이터분석에서는 주로 inner 는 outter 함수 안에서 사용한다)
        print("inner")

    return inner

f = outer()

f()

outer()()
```

```python
def plus(x):
    return x + 10

def minus(x):
    return x - 10

funcs = [plus, minus]  # 함수들을 리스트화 시켜서 사용

print(funcs[0](100))
print(funcs[1](100))
```
