```python
s = {1, 2, 3}

print(len(s)) #개수
print(2 in s) #포함 여부

s.clear() #전체 삭제
print(s) #set()
```

```python
# 패킹, 언패킹시 * 연산자
# a, b = [1, 2, 3, 4] too many values to unpack
*a, b = [1, 2, 3, 4]
print(a)
print(b)

c, *d = (1, 2, 3, 4)
print(c)
print(d)

e, *f, g = (1, 2, 3, 4)
print(e)
print(f)
print(g)

h = [1, 2, 3]
print(h)
print(*h) # print(h[0], h[1], h[2])
```
