```python
a = [1, 2, 3]
b = [4, 5]
result = a + b
print(result) # [1, 2, 3, 4, 5]
print(a)
print(b)

a = a + b
print(a)

c = "py"
c = c + "thon"
print(c)
```

```python
nums = [10, 20]
result = nums * 3
print(result) # [10, 20, 10, 20, 10, 20]

# [30, 60] Numpy 라이브러리로 행렬객체를 만들어 사용
```

```python
nums = [10, 20, 30, 40, 50]

print(nums[1:4]) # [20, 30, 40]
print(nums[:3]) # [10, 20, 30]
print(nums[2:]) # [30, 40, 50]
print(nums) #  [10, 20, 30, 40, 50]
```

```python
a = [10, 20, 30]
b = a

print(a)
print(b)
print(a == b)
print(a is b)

b[0] = 999
print(a) 
print(b) 
```

```python
a = [1, 2, 3]
b = [1, 2, 3]
c = [1, 2, 4]

print(a == b) # True
print(a == c) # False
```

```python
# 정렬 메서드
nums = [50, 10, 40, 20]
n2 = nums.sort() # 오름차순
print(nums)
print(n2)
```

```python
nums = [50, 10, 40, 20]
nums.sort(reverse=True) # 내림차순
print(nums)
```
