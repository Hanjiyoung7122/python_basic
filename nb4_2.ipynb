```python
a = 10
b = a

print(a, b) # 10 10
print(a is b)  # True 같은 객체 참조
```

```python
s1 = "Python"
s2 = s1

print(s1 is s2) # True
print(s1 == s2) # True
```

```python
nums1 = [10, 20, 30]
nums2 = nums1

print(nums1 is nums2) # True
print(nums1 == nums2) # True
```

```python
nums1 = [10, 20, 30]
nums2 = [10, 20, 30]

print(nums1 is nums2) # False 가변데이터이기 떄문에 공유참조 불가
print(nums1 == nums2) # True
print(id(a))
print(id(b))
```

```python
a = [1, 2, 3, 4]
b = a

print(a == b) # True
print(a is b) # True

# 내장 함수 id()를 이용하면 동일 객체를 참조하는지 확인 가능
print(id(a))
print(id(b))
```
