```python
# with문 사용하지 않는 경우
try:
    f = open("students.csv", "r", encoding="utf-8")
    print(f.read())
finally:
    f.close()   # 직접 닫아야 함. 실수로 빼먹거나, 실행전 에러가 발생하면 메모리 누수가 발생한다. 
```

```python
# with문을 사용하는 경우
with open("students.csv", "r", encoding="utf-8") as f:
    print(f.read())
```

```python
with open("students.txt", "r", encoding="utf-8") as f:
    # 한 줄씩 읽기.
    for line in f: 
        print(line.strip()) # 라인 끝의 줄바꿈이 문자 포함되어 두번 줄바꿈이 일어난다(print함수도 줄바꿈이 기본값)
```

```python
# 한 라인을 구분자","를 이용하여 두개의 문자열 값으로 저장
names = []
scores = []
with open("students.txt", "r", encoding="utf-8") as f:
    for line in f:
        name, score = line.strip().split(",")
        #print(name, int(score))
        names.append(name)
        scores.append(score)


print(names)
print(scores)

```

```python

s = [] # [(), (), ()]

with open("students.txt", "r", encoding="utf-8") as f:
    for line in f:
        a = line.strip().split(",")
        #print(name, int(score))
        s.append(tuple(a))

print(s)
```

```python
import csv

# open()은 내장함수
with open("students.csv", "r", encoding="utf-8") as f:
    reader = csv.reader(f)
    print(type(reader))
    for row in reader:
        print(row)
        print(type(row)) # <class "list">
```

```python
with open("students.csv", "r", encoding="utf-8") as f:
    reader = csv.reader(f)

    next(reader)

    for row in reader:
        print(row)
```

```python
with open("students.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f) # 헤드라인은 키, 한줄의 내용은 값

    for row in reader:
        print(row)
```

```python
with open("students.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)

    data = list(reader)

print(data)

data[2][" 영어"] = 80

print(data[2])
```

```python
with open("students_result.csv", "w", newline="", encoding="utf-8-sig") as f:
    # data[0].keys() → 첫 번째 딕셔너리의 키(컬럼명) 사용    
    writer = csv.DictWriter(f, fieldnames=data[0].keys())
    # 헤더(컬럼명) 저장
    writer.writeheader()   
    # data 리스트에 있는 모든 딕셔너리를 한 번에 저장
    writer.writerows(data)  

print("CSV 저장 완료")
```
