```python
def calc(a, b):
    x = a + b
    y = a - b
    z = a * b
    return x , y, z # 3개의 값을 반환하는 것이 아니라 tuple로 (x, y, z) 묶어서 반환된다

result2 = calc(10, 5)
print(result2)

r1, r2, r3 = calc(10, 5) # 언패킹
print(r1)
print(r2)
print(r3)
```

```python
def fun1(a, b):
    print("매개변수 a의 인수: ", a)
    print("매개변수 b의 인수: ", b)

# 함수 호출시 인자값(인수)의 위치로 매개변수 값 결정
fun1(1, 2)
fun1(10, 20)

# 함수 호출시 매개변수 이름을 지정하여 인자값(인수) 입력
fun1(a=100, b=200)
fun1(b=777, a=999) # 매개변수 이름을 지정하여 인수 입력시(키워드 인수) 순서는 상관없음.
```

```python
def my_intersect(a, b) -> list:
    """
    
    a: 리스트
    b: 리스트
    반환타입: 리스트
    내용: 두개의 리스트에 중복되는 값만을 새로운 리스트로 반환
    """
    result = []
    for e in a:
        if e in b:
            result.append(e)
    return result


a = [1, 2, 3, 4, 5, 6, 7, 78]
b = [5, 6, 7, 10 , 29, 111]
r1 = my_intersect(a, b)
print(r1)
help(my_intersect)
```

```python
# 두 시퀀스의 공통 요소 개수

def my_count_common(a, b):
    count = 0
    for e in a:
        if e in b:
            count += 1
    return count


print(my_count_common([1, 2, 3], [2, 3, 5]))
print(my_count_common((1, 2, 3), (2, 3, 5)))
print(my_count_common({1, 2, 3}, {2, 3, 5}))
print(my_count_common({"id":"abc", "sn":"1111"}, {"id":"xyz", "pw":"1234"}))
```

```python
# a에는 있고 b에는 없는 요소

def my_difference(a, b):
    result = []
    for e in a:
        if e not in b:
            result.append(e)
    return result


print(my_difference([1, 2, 3], [2, 3, 5]))
print(my_difference((1, 2, 3), (2, 3, 5)))
print(my_difference({1, 2, 3}, {2, 3, 5}))
print(my_difference({"id":"abc", "sn":"1111"}, {"id":"xyz", "pw":"1234"}))
```

```python
# 두 시퀀스의 합집합

def my_union(a, b):
    result = []

    for e in a:
        if e not in result:
            result.append(e)

    for e in b:
        if e not in result:
            result.append(e)

    return result


print(my_union([1, 2, 3], [2, 3, 5]))
print(my_union((1, 2, 3), (2, 3, 5)))
print(my_union({1, 2, 3}, {2, 3, 5}))
```

```python
def my_count(data, target):

    count = 0

    for e in data:
        if e == target:
            count += 1

    return count


print(my_count([1, 2, 1, 3, 1], 1))
print(my_count(("a", "b", "a"), "a"))
print(my_count("banana", "a"))
```

```python
# 중복 제거

def my_unique(data):

    result = []

    for e in data:
        if e not in result:
            result.append(e)

    return result


print(my_unique([1, 2, 2, 3, 1, 4]))
print(my_unique(("A", "A", "B", "C")))
print(my_unique("banana"))
```
