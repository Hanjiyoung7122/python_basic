```python
age = 20
result = ""
if age >= 19:
    result = "성인"
else:
    result = "미성년자"
print(result)
```

```python
num = 7

if num % 2 == 0:
    print("짝수")
else:
    print("홀수")
```

```python
scores = [95, 60, 80, 45]

for score in scores: # x4
    if score >= 70:
        print("합격")
    else:
        print("불합격")


numbers = [3, 8, 11, 20, 25]

for num in numbers:
    if num % 2 == 0:
        print(f"{num} : 짝수")
    else:
        print(f"{num} : 홀수")
```
